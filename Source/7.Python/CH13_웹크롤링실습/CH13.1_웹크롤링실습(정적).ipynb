## yes24의 베스트셀러 정보를 제공하는 사이트에서 베스트셀러 정보를수집해서 파일에 저장하세요.
- 베스트셀러 정보 수집 주소 : http://www.yes24.com/24/category/bestseller

```python
from urllib import request
import pandas as pd
from bs4 import BeautifulSoup

def load_url(url):
  with request.urlopen(url) as f:
    charset = f.headers.get_content_charset()
    html = f.read().decode(charset)
  return BeautifulSoup(html, 'html5lib')

def to_int(raw):
  nums = raw.split("(")[0].replace('원', '').replace(',', '')
  return int(nums)

bestseller_url = "http://www.yes24.com/24/category/bestseller"
bs = load_url(bestseller_url)
books = bs.select('#bestList > ol > li')
rank = [b.select_one('p:nth-of-type(3) a').text for b in books]
titles = [b.select_one('p:nth-of-type(3) a').text for b in books]
writer = [b.select_one('p.aupu').text for b in books]
prices = [to_int(b.select_one('p.price').text) for b in books]

df = pd.DataFrame(
    {'책이름': titles, '저자 및 출판사':writer, '가격': prices},
    columns=['책이름','저자 및 출판사', '가격']
)
df
```

```python
df.to_csv("1_yes24.csv")
```
